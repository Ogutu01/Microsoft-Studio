## Data Preparation
Data preparation is a crucial stage in this project for a number of reasons, including:

* Data Quality Assurance: Ensuring data quality is crucial. Errors, missing values, and inconsistent data are all possible. We can find and fix these problems by carefully gathering the data, ensuring the dependability of our analysis.

* Data Integration: We are utilizing a number of datasets from various sources. By addressing differences in variable names, units, and formats, data preparation enables us to harmonize these datasets, allowing for meaningful cross-referencing and analysis.

* Feature engineering: New features might be developed or current ones modified in order to improve analysis.

* Handling Missing Data: Analysis results can be greatly impacted by missing data. In order to achieve a robust analysis, handling missing values must be decided, whether through imputation, deletion, or other suitable approaches.

* Outlier detection: For statistical validity, outliers must be found and dealt with. We can use methods like visual inspection or statistical testing to find outliers and handle them correctly with the help of data preparation.



We will drop the columns that we do not require for the analysis.

```python
# Dropping unnecessary columns from the DataFrame

# Creating a list of columns to drop
cols_to_remove = ["domestic_gross_x","foreign_gross", "year","id_x","original_title", "release_date_x", "id_y", "release_date_y", "movie"]
cols = [col for col in merged_df if col not in cols_to_remove]
merged_df = merged_df[cols]

merged_df.head() # This results in a shape of shape (1395, 10)
```

We again check to see if there are any missing values in the specific columns in the DataFrame. We will also need to deal with them in order to come up with a strategy. We have three options for dealing with missing values removing them from the dataset, keeping them, or replacing them with another value.

```python
# Checking for any missing values
merged_df.isna().any()
```

```python
# Checking the number of missing values in the respective columns
merged_df.isna().sum() # Found 1 in the "studio" column
```

Since there is only one missing value in the studio column, we will remove the offending row.

```python
# Dropping the offending row
merged_df = merged_df.dropna(subset = ["studio"])
merged_df.isna().sum()
merged_df.shape # This results in a shape of (1394, 10)
```

We check for duplicates, remove them then recheck again to ensure that duplicates no longer exist.

```python
# Checking for duplicate values
merged_df.duplicated().value_counts()
```

```python
# Dropping the duplicate values 
merged_df = merged_df.drop_duplicates()
merged_df.shape # This results in a shape of (1270, 10)
```

```python
# Rechecking to ensure the duplicate values no longer exist
merged_df.duplicated().value_counts()
```

Data preparation and analysis need the conversion of different data types, known as data casting. It entails converting a variable's data type from one format to another (for example, from string to integer or float). Data casting is significant because it can guarantee data accuracy, simplify analysis, and provide insightful information. In the next cwlls we will figure out the columns that require data casting by first checking the DataFrame's info.

```python
df_info(merged_df)
```

Fron the info, we need to convert three columns, namely: `production_budget`, `domestic_gross_y` and `worldwide_gross` from object type to float data types using the `astype()` function.We will also remove the dollar signs and commas in the `domestic_gross_y` and `worldwide_gross` columns by using `str.replace()`.

```python
# Create a list of columns to clean and convert
clean_and_convert = ["production_budget", "domestic_gross_y", "worldwide_gross"]

# Remove dollar signs and commas, and convert to float for each column
for column in clean_and_convert:
    merged_df[column] = merged_df[column].str.replace("$", "").str.replace(",", "").astype(float)

merged_df.head()
```

```python
df_info(merged_df) # Checking to see the data casted columns
```

The dataset is now clean and ready for analysis. To improve the data, several crucial actions were conducted. A single missing value in the'studio' column was quickly deleted when we meticulously used the `.isna()` method to check for missing values. As a consequence, we were left with a dataset that had the shape (1394, 10). The next crucial step was to use the `.duplicated()` method to find and handle duplicates. This revealed 124 duplicated values that were immediately removed, giving us a DataFrame with the shape (1270, 10). We data casted the `production_budget`, `domestic_gross_y`, and `worldwide_gross` columns to float data type to guarantee numerical consistency. Additionally, we used the `.str.replace()` method to remove dollar signs and commas from these columns and the `.astype()` method to convert the data types. These thorough data cleaning procedures have now prepared the dataset for intelligent analysis.

### Problem Statement 1: To profile the competitive landscape within the movie industry, including the performance of key players
This inquiry aims to analyze the competitive dynamics of the film business. To comprehend their respective performances, it entails a detailed research of the studios. Microsoft can make strategic decisions about prospective alliances, collaborations, or areas of opportunity to build a solid footing in this competitive sector by acquiring knowledge into the competitive forces at play.

```python
# Creating a copy of the merged dataset
studio_ratings_df = merged_df.copy()
studio_ratings_df.head()
```

```python
studio_ratings_df["studio"].value_counts()
```

Using bivariate analysis we plot a graph to check the dominance of the studios based on the number of movies they have released.

```python
studio_movie_counts = studio_ratings_df['studio'].value_counts()

# Create a bar chart
figure, ax = plt.subplots(figsize =(8, 14))
sns.barplot(x=studio_movie_counts.values, y=studio_movie_counts.index, palette='viridis')
ax.set_xlabel('Number of movies')
ax.set_ylabel('Studio Name')
ax.set_title('Number of Movies Released by Studio')
plt.savefig("Number of Movies vs. Studio Name.png")
```

```python
# Taking 'production_budget' and 'worldwide_gross' as the columns for budget and gross earnings
studio_success = studio_ratings_df.groupby('studio').agg({
    "production_budget" : "sum",   # Calculate total worldwide gross earnings
    "worldwide_gross": "sum" # Calculate total production budget spendings
})      

# Print success metrics
studio_success = studio_success.sort_values(by = "worldwide_gross", ascending = False).reset_index()
studio_success
```

Next we examine the top studios based on worldwide gross and identify the dominant players in the industry. This helps in understanding which studios have consistently delivered successful movies.

```python
# Dropping rows that have studios operating at a loss
gross_below = studio_success['production_budget'] > studio_success['worldwide_gross']
studio_success = studio_success.drop(studio_success[gross_below].index).reset_index(drop = True)
studio_success
```

Analyzing the distribution of worldwide gross among the top 20 studios can reveal how market share is distributed. You can identify if a few studios hold a significant portion of the market or if it's relatively evenly distributed.

```python
# Create a bar plot to show the aove analysis
figure, ax = plt.subplots(figsize=(10, 10))
sns.barplot(x = studio_success['worldwide_gross'].head(20), y = studio_success['studio'].head(20))
ax.set_xlabel('Total Worldwide Gross Earnings')
ax.set_ylabel('Studio Name')
ax.set_title('Studio Success Metrics')
plt.savefig("Studio Name vs. Success Metrics.png")
```

The above analysis arises the following insights:
* Competitive Benchmarking: A corporation can evaluate its position in the market by comparing its performance to that of its rivals through the strategic process known as competitive benchmarking. This analysis entails the identification of immediate competitors, the gathering and analysis of information, including financial metrics, and market share. Companies use this assessment to uncover best practices used by top competitors, identify strengths and weaknesses, create performance goals, execute changes, and continuously track progress. In the end, this iterative approach helps firms thrive in a competitive environment by fostering performance improvement, informed decision-making, market positioning and efficient strategy planning.To Microsoft, as a new company entering the industry, this analysis provides a benchmark to compare against. It helps in setting realistic goals for market share and revenue targets.

### Problem Statement 2: To assess the success of different languages based on their worldwide gross earnings
This analysis tries to reveal the importance of languages in the current box office scene. We want to determine which languages now control the global by comparing the global gross revenues of films released in a variety of languages. By taking into account audience preferences and enhancing global appeal, Microsoft will be able to make educated decisions about the preferred languages for its film productions.

```python
# Creating a copy of the merged dataset.
languages_df = merged_df.copy()
languages_df.head()
```

```python
languages_df["original_language"].value_counts()
```

Using the `.map()`method on a pandas Series with a dictionary of values to replace the values in the `original_language` column to desired values that can be easily understood.

```python
# Creating a dictionary with the abbreviations as keys and their respective full names as values.

language_mapping = {
   "ar" : "Arabic",
    "da" : "Danish",
    "de" : "German",
    "el" : "Greek",
    "en" : "English",
    "es" : "Spanish",
    "fa" : "Persian",
    "fr" : "French",
    "he" : "Hebrew",
    "hi" : "Hindi",
    "hu" : "Hungarian",
    "ko" : "Korean",
    "pl" : "Polish",
    "ru" : "Russian",
    "te" : "Telugu",
    "th" : "Thai",
    "tr" : "Turkish",
    "xx" : "Unknown"
}
```

```python
# Call the .map() method to return a Series with the abbreviations transformed into full names:
languages_df["original_language"].map(language_mapping)
```

```python
# Replace the "original_language" column in languages_df with these new, transformed values:
languages_df["original_language"] = languages_df["original_language"].map(language_mapping)
languages_df["original_language"].value_counts()
```

We plot a bar graph to check the top 5 most common languages used based on the number of movies made in that language.

```python
language_counts = languages_df["original_language"].value_counts()

# Plot a bar graph of the five most common languages
top_languages = language_counts.head()
figure, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x = top_languages.index, y = top_languages.values, ax=ax)
ax.set_xlabel("Languages")
ax.set_ylabel("Number of Movies")
ax.set_title("Top 5 Most Common Languages by Number of Movies")
plt.savefig("Number of Movies vs. Languages.png")
```

The top 5 most common languages are, English as the lead, French, Hindi, Spanish and Russian.

Next we evaluate the `worldwide_gross` associated by each of the top 5 most common languages and analayze which one produces high gross values.

```python
# Calculate the top 5 most common languages based on count
x = languages_df["original_language"].value_counts().head().index
y = languages_df.groupby('original_language')['worldwide_gross'].mean().loc[x]

# Create a DataFrame for plotting
data_for_plot = pd.DataFrame({'Languages': x, 'Worldwide Gross Revenue': y})

# Plot a bar graph of the five most common languages
figure, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=data_for_plot, x='Languages', y='Worldwide Gross Revenue', ax=ax)
ax.set_xlabel("Top 5 Languages")
ax.set_ylabel("Worldwide Gross Revenue")
ax.set_title("Top Multilingual Languages by Average Worldwide Gross Revenue")
plt.savefig("Top Multilingual Languages vs. Average Worldwide Gross Revenue.png")
```

When compared to films in other languages, the study of the movie dataset shows that English-language films regularly have the greatest worldwide box office take. Several things can be used to support this conclusion. First off, as English is one of the most extensively used languages in the world, a larger audience for English-language films is possible. In addition, many English-language films are produced by renowned studios, which can increase their marketability and appeal to a broad audience.
Therefore, Microsoft should consider producing a significant portion of its movie portfolio in English. This aligns with the global appeal of English-language cinema and its potential to reach a broader audience.

### Problem Statement 3: To assess the financial implications of production budgets on the performance of movies
This analysis explores the financial aspects of the film industry, focusing on the connection between production costs and total film quality. by looking at how spending on a film's budget affects its commercial success. These data will enable Microsoft's film division to allocate resources intelligently, assuring the best possible financial results while creating high-caliber cinematic experiences.

```python
# Creating aa copy of the merged dataset
budgets_df = merged_df.copy()
budgets_df.head()
```

The aim is to plot scatter plots which are crucial for data-driven decision-making in the movie industry. They provide insights into how financial investments in production budgets correlate with revenue generation and how domestic and global markets interact. This information informs budgeting, marketing, and distribution strategies, ultimately contributing to the financial success of movie projects.

* Worldwide Gross vs. Production Budget: To establish whether there is a positive correlation, a scatter plot of the worldwide gross (on the y-axis) against the production budget (on the x-axis) is used. A positive association exists between production budget and worldwide gross if the majority of the data points cluster from left to right in an upward direction. Making decisions on how to allocate funds for film production requires this understanding.


* Domestic Gross vs. Worldwide Gross: A scatter plot of domestic gross (on the x-axis) and worldwide gross (on the y-axis) shows how domestic and international revenues are related. If there is a significant positive correlation, it may mean that films that do well domestically also frequently do well internationally, which could have an impact on distribution and marketing plans.


* Domestic Gross vs. Production Budget: To determine the effect of the production budget on domestic earnings, a scatter plot of the production budget (on the x-axis) against domestic gross (on the y-axis) is used. A positive association would suggest that larger spending and bigger domestic gross are related, informing marketing and target audience decisions.

```python
# Create subplots for domestic gross and worldwide gross

# Domestic Gross vs. Production Budget
fig, ax1= plt.subplots(figsize=(10, 6))

sns.scatterplot(data = budgets_df, x = "production_budget", y = "domestic_gross_y")
ax1.set_xlabel("Production Budget")
ax1.set_ylabel("Domestic Gross")
ax1.set_title("Production Budget vs. Domestic Gross")
plt.savefig("Production Budget vs. Domestic Gross.png")

# Worlwide Gross vs. Production Budget
fig, ax2 = plt.subplots(figsize=(10, 6))

sns.scatterplot(data = budgets_df, x = "production_budget", y = "worldwide_gross")
ax2.set_xlabel("Production Budget")
ax2.set_ylabel("Worldwide Gross")
ax2.set_title("Production Budget vs. Worldwide Gross")
plt.savefig("Production Budget vs. Worldwide Gross.png")

# Domestic Gross vs. Worldwide Gross
fig, ax3 = plt.subplots(figsize = (10, 6))
sns.scatterplot(data = budgets_df, x = "domestic_gross_y", y = "worldwide_gross")
ax3.set_xlabel("Domestic Gross")
ax3.set_ylabel("Worldwide Gross")
ax3.set_title("Domestic Gross vs. Worldwide Gross")
plt.savefig("Domestic Gross vs. Worldwide Gross.png")
```

```python
# Creating a function that returns the correlation coefficients between the specified columns
def correlation (x, y):
    corr = np.corrcoef(x, y)
    return corr[0][1]
```

```python
# Calculate the correlation coefficients
# To quantitatively measure the strength and direction of the relationship between production budgets and gross earnings. 
# A positive correlation suggests that higher budgets lead to higher earnings.
x = budgets_df["production_budget"]
y = budgets_df["domestic_gross_y"]

x1 = budgets_df["production_budget"]
y1 = budgets_df["worldwide_gross"]

x3 = budgets_df["domestic_gross_y"]
y3 = budgets_df["worldwide_gross"]

print("The correlation between Production budget and Domestic gross is: ", correlation(x, y))
print("The correlation between Production budget and Worldwide gross is: ", correlation(x1, y1))
print("The correlation between Production budget and Worldwide gross is: ", correlation(x3, y3))
```

* Domestic Gross and Production Budget: There exists a moderately strong positive relationship between the `production_budget` invested in a movie and its `domestic_gross_y` earnings. This finding suggests that allocating higher budgets to movie production is associated with a greater likelihood of achieving higher earnings in the domestic market.

* Worldwide Gross and Production Budget: Similarly, the positive correlation between `worldwide_gross` and `production_budget` indicates that higher production budgets are linked to increased worldwide gross earnings. This emphasizes the financial impact of strategic budget allocation in movie production.

* Domestic Gross and Worldwide Gross: The notably high correlation between `domestic_gross_y` and `worldwide_gross` highlights that movies performing well domestically tend to perform exceptionally well on the global stage. This underlines the interconnected nature of these markets and the potential to leverage domestic success for international appeal.

These strong positive connections highlight the financial significance of production budget allocation in the film business, to sum up. In light of the fact that wise production investments can have a major impact on both domestic and international profitability, Microsoft's new film studio should take these findings into account when deciding how much money to allocate to filmmaking. Recognizing the interdependence of the domestic and international markets further highlights the importance of flexible distribution and marketing tactics for maximizing profit potential.
